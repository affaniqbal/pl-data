## Import packages

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy import stats
```

## Load data using numpy

```python
# Use numpy.loadtxt to import our data
x_data, y_data = np.loadtxt('PL_data.csv', unpack=True, delimiter=',')
```

## Use Scipy to describe data

```python
stats.describe(x_data)
```

```python
stats.describe(y_data)
```

## Quickly visualise data

```python
# Create a figure object and an axis object
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

# Build on the figure object by plotting the x,y data 
ax.plot(x_data, y_data)

# Use the show function to print the figure object
plt.show()
```

## Plotting and saving a publication-ready 

```python
# Create figure object and store it in a variable called 'fig'
fig = plt.figure(figsize=(4, 4))

# Add axes object to our figure that takes up entire figure
ax = fig.add_axes([0, 0, 1, 1])

# Edit the x axis ticks
ax.xaxis.set_tick_params(which='major', size=10, width=2, direction='in', labelsize = 16)
ax.xaxis.set_major_locator(ticker.MultipleLocator(50))

# Edit the y axis ticks
ax.yaxis.set_tick_params(which='major', size=10, width=2, direction='in', labelsize = 16)
ax.yaxis.set_major_locator(ticker.MultipleLocator(1000))

# Plot the data and add linewidth and label
ax.plot(x_data, y_data, linewidth=3,  label='Control sample')

# Add the x and y-axis labels
ax.set_xlabel('Wavelength (nm)', labelpad=10, fontsize = 16)
ax.set_ylabel('PL intensity (counts)', labelpad=10, fontsize = 16)


# Set the x-axis limits
ax.set_xlim(690, 900)

# Set the y-axis limits
ax.set_ylim(0,3700)


# Add legend to plot
ax.legend(bbox_to_anchor=(1, 1), loc=1, frameon=False, fontsize = 16)

# Save figure
plt.savefig('Final_Plot.png', dpi=600, bbox_inches='tight')

# Show figure
plt.show()
```
